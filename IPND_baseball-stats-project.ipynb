<h3>Data Wrangling</h3>

<pre>We will be looking at 2 tables - Teams and Salaries
In order for me to have some consistent data,
we looked at all seasons with around 162 games, > 1960 season</pre>

```python
# teams_full_df = teams_df[teams_df['yearID'] >= 1961]

# Above code changed to a function - truncate_data_yearID

# Function: truncate_data_yearID
# Input: df as dataframe
#        year as numeric year to slice from
# Output: dataframe without data less than year input

def truncate_data_yearID(df, year):
    return df[df['yearID'] >= year]

teams_full_df = truncate_data_yearID(teams_df, 1961)

teams_full_df

# Visually checked to see that the data has yearID from 1961 to 2015
```

<pre>[1] Checking the salaries table, we see salaries information start >= 1985
[2] Also, the data is for individual player salaries</pre>

```python
salaries_df.head()
```

<pre>Here, we grouped the salaries by teamID to address above. [2]</pre>

```python
team_salaries_df = salaries_df.groupby(['yearID', 'teamID']) \
                   .sum().reset_index()
print team_salaries_df.head()
```

<pre>We changed the teams table, narrowing the years >= 1985 to address above [1]
Used function truncate_data_yearID</pre>

```python
teams_full_df = truncate_data_yearID(teams_df, 1985)

teams_full_df.head()
```

<pre>Joined the two tables - teams_full_df and team_salaries_df
result_df will be the main dataframe we will be using for analysis</pre>

```python
result_df = team_salaries_df.merge(teams_full_df,
                                   how='inner',
                                   on=['yearID', 'teamID']
                                  )
result_df.head()
```

<pre>Not all columns are needed
Looking at the basic games, wins, losses columns with team salaries</pre>

```python
# Function: data_pivot_games
# Input: df as dataframe
# Output: pivot table with W, L, G, salary columns
#         indexed by yearID and teamID

def data_pivot_games(df):
    return df.pivot_table(values=['L', 'W', 'G', 'salary'],
                          index=['yearID', 'teamID'],
                          aggfunc=np.sum
                          )

games_overall_pv = data_pivot_games(result_df)
games_overall_pv.head()
```

```python
# Function: col_name_change
# Input: df as data
# Output: df with column names changed
# Notes: can always add more column name to change

def col_name_change(df):
    return df.rename(columns={'G': 'No_of_Games',
                              'L': 'No_of_Losses',
                              'W': 'No_of_Wins',
                              'H': 'No_of_Hits',
                              'R': 'Runs_Scored',
                              'HR': 'Homeruns',
                              'SB': 'Stolen_Bases',
                              'salary': 'Team_Salaries'
                             }
                    )

games_overall_pv = col_name_change(games_overall_pv)
```

We see that there are no missing values in games_overall_pv</pre>

```python
print games_overall_pv.isnull().sum()
```

<pre>Here is the same information as the pivot table in dataframe
Not indexed by yearID and teamID</pre>

```python
games_overall_df = result_df[['yearID','teamID','G','W','L','salary']]
games_overall_df.head()
```

<pre>Again, if we need the column names changed, we use function

```python
games_overall_df = col_name_change(games_overall_df)
games_overall_df.head()
```

```python
# Function: compare_grid
# Input: pv as pivot table
# Ouput: various pairplots in reference to each of the columns in pivot table

def compare_grid(pv, title_desc):
    g = sns.PairGrid(pv)
    plt.subplots_adjust(top=0.95)
    sns.plt.suptitle(title_desc)
    g.map_diag(plt.hist)
    g.map_offdiag(plt.scatter)
    return

compare_grid(games_overall_pv,
             'Grid Pair Plots of Games, Wins, Losses, and Team Salaries')
```

<pre>Here is a function that will standardize data</pre>

```python
# Function: standardize_data
# Input: x as data
# Output: x_axis as data
# Notes: used level='yearID' to calculate over subset by yearID,
#        not all population

def standardize_data(x):
    x_axis = (x - x.mean(level='yearID')) / x.std(ddof=0,level='yearID')
    return x_axis
```

```python
# Function: compare_linear with lmplot
# Input: df as dataframe
#        x_axis as column name for x
#        y_axis as column name for y
# Output: lmplot of data
# Notes: Can use dataframes or pivot tables

def compare_linear(df, x_axis, y_axis, title_desc):
    g = sns.lmplot(x=x_axis, y=y_axis, data=df)
    plt.subplots_adjust(top=0.92)
    sns.plt.suptitle(title_desc)
    return

compare_linear(result_subset_df, 'Team_Salaries', 'No_of_Wins',
               'Linear Regression Plot - Team Salaries vs Number of Wins')

# Here we see a positive relationship of # of wins as team salaries increases
```

<pre>For reference, we looked at various scatter plots with
relationships of team salaries and wins.
Some data were converted to using standardize_data function.</pre>

```python
# Function: plot_xy
# Inputs: df as data
#         x_col is name of x-axis column
#         y_col is name of y-axis column
#         graph_title is Title of Plot
# Output: Scatter plot of inputs


def plot_xy(df, x_col, y_col, graph_title):
    x_axis = df[x_col]
    y_axis = df[y_col]
    plt.title(graph_title)
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.scatter(x_axis, y_axis)
    return

plot_xy(games_overall_pv, 'Team_Salaries', 'No_of_Wins',
        'Scatter Plot of Wins Per Team Salaries')
```

<pre>The above scatter plot shows the that have been spending
more than $180 had a winning season (more than 81 wins).
</pre>

```python
# Function: plot_x_std
# Inputs: df as data
#         x_col is name of x-axis column, std dev calc using function
#         y_col is name of y-axis column
#         graph_title is Title of Plot
# Output: Scatter plot of inputs

def plot_x_std(df, x_col, y_col, graph_title):
    df_std = standardize_data(df)
    x_axis = df_std[x_col]
    y_axis = df[y_col]
    plt.title(graph_title)
    plt.xlabel(x_col + ' Std Dev')
    plt.ylabel(y_col)
    plt.scatter(x_axis, y_axis)
    return

plot_x_std(games_overall_pv, 'Team_Salaries', 'No_of_Wins',
           'Scatter Plot of Wins Per Team Salaries (Std Dev)')
```

<pre>The above scatter plot has the Team Salaries standardized.
We see that teams with team salaries above 2 standard deviations
have more wins, but shows that there are few teams that greatly
outspends than others.
Also, we see that teams who tend to spend far less, will in most
cases have a losing season (less than -2 std dev of team salaries).
</pre>

```python
# Function: plot_y_std
# Inputs: df as data
#         x_col is name of x-axis column
#         y_col is name of y-axis column, std dev calc using function
#         graph_title is Title of Plot
# Output: Scatter plot of inputs

def plot_y_std(df, x_col, y_col, graph_title):
    df_std = standardize_data(df)
    x_axis = df[x_col]
    y_axis = df_std[y_col]
    plt.title(graph_title)
    plt.xlabel(x_col)
    plt.ylabel(y_col + ' Std Dev')
    plt.scatter(x_axis, y_axis)
    return

plot_y_std(games_overall_pv, 'Team_Salaries', 'No_of_Wins',
           'Scatter Plot of Wins (Std Dev) Per Team Salaries')
```

<pre>The above scatter plot shows Number of Wins standardized.
We see that majority of of wins will be in the range from 2 to -2
standard deviations. Those teams with wins aboe 2 std dev and below
-2 std dev, may indicate that the early seasons after 1985, salaries
may have not been a factor in number of wins.
However, with some team salaries reaching above $150 million in 
recent years, there is a positive effect where teams win more
above 1 std dev, and not be below -1 std dev.</pre>

```python
# Function: plot_xy_std
# Inputs: df as data
#         x_col is name of x-axis column, std dev calc using function
#         y_col is name of y-axis column, std dev calc using function
#         graph_title is Title of Plot
# Output: Scatter plot of inputs

def plot_xy_std(df, x_col, y_col, graph_title):
    df_std = standardize_data(df)
    x_axis = df_std[x_col]
    y_axis = df_std[y_col]
    plt.title(graph_title)
    plt.xlabel(x_col + ' Std Dev')
    plt.ylabel(y_col + ' Std Dev')
    plt.scatter(x_axis, y_axis)
    return

plot_xy_std(games_overall_pv, 'Team_Salaries', 'No_of_Wins',
           'Scatter Plot of Wins (Std Dev) Per Team Salaries (Std Dev)')
```

<pre>The above scatter plot has both Team Salaries and Number of Wins
standardized. We see that majority of the data will be within 2 standard
deviations.
Teams that far outspend above 2 std dev will not greatly impact number
of wins above 2 std devs. This means that although teams may far outspend
their peers, they will not dominate completely.</pre>

<pre>Next, for reference we looked at various stats of data.</pre>

```python
# Function: describe_data_yearID
# Input: df as data
# Output: by yearID, print of mean, std, min, max
# Notes: Use pivot tables

def describe_data_yearID(df):
    print 'Mean:'
    print df.mean(level='yearID')
    print ''
    print 'Standard Deviation:'
    print df.std(level='yearID')
    print ''
    print 'Minimum:'
    print df.min(level='yearID')
    print ''
    print 'Maximum:'
    print df.max(level='yearID')
    return

describe_data_yearID(games_overall_pv)
```

of the teams will have wins in the range of 70 to 90 games,
while those teams with high salaries will have 70 to 100 wins
(std dev range of -1 to 2 as shown in
'Scatter Plot of Wins (Std Dev) Per Team Salaries (Std Dev)')

and maximum wins fall around 100 (2 std dev)

<h4>Limitations</h4>
<pre>We do have some limitations with the analysis. The first limitation is 
inflation. Inflation rate has been high in the 1980's, where inflation rate 
has been low in the 2010's. In order to compare team salaries of each year, 
it may be interesting to see salaries adjusted for inflation. This would 
normalize the data and provide more accurate analysis when looking at team 
salaries.

Second limitation is that the salaries was dependant upon one source. As we
have seen, two data points were not accurate and had to be adjusted. We may
overcome this limitation by looking at other sources of information and
compare for salary accuracy.

Third limitation is that players often get traded during the season, sometimes
late in the season. This affects salary information, and it is uncertain how 
the data captures trades. We will have to delve deeper and see if trades are 
actually captured and how they are treated in the data.
